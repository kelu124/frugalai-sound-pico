```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

# Exploring what "chainsaw" looks like

```python
z = np.load("spectrograms/v5_create_arm_spectrogram_for_map_label0.npz")
signal = z["signal"]
label = z["label"]

for x in range(len(signal)):
    avg =  np.mean(signal[x][:,35:50]) 
    if not x:
        S1 = signal[0] / avg
    else:
        S1 += signal[x] / avg
plt.imshow(np.transpose(S1))
```

# MELs

```python
z = np.load("spectrograms/v5_create_mel_spectrogram_for_map_label0.npz")
signal = z["signal"]
label = z["label"]

for x in range(len(signal)):
    avg =  np.mean(signal[x][:,35:50]) 
    if not x:
        S1 = signal[0] / avg
    else:
        S1 += signal[x] / avg


fig, ax = plt.subplots(figsize=(10,6))

plt.imshow(np.transpose(S1), aspect=1.5)

ax.set(xticks=np.arange(0, 280-1)[::50], xticklabels=[int(x*3/28)/10 for x in np.arange(0, 280-1)][::50]);
ax.set(yticks=np.arange(0, 128)[::30], yticklabels=[int(x*8/12.8)/10 for x in np.arange(0, 128)][::30]);
plt.xlabel("Time (s)")
plt.ylabel("Freq (kHz)")

plt.title("Presence of chainsaws (mel_spectrogram)\n ")
```
